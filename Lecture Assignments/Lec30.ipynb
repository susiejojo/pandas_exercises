```python
import pandas as pd
```

```python
train=pd.read_csv("http://bit.ly/kaggletrain")
```

```python
train.head()
```

```python
train['Sex_num']=train.Sex.map({'female':1,'male':0})
```

```python
train.head()
```

```python
train['Name_len']=train.Name.apply(len)
```

```python
train.head()
```

```python
import numpy as np
```

```python
train["Fare_ceil"]=train.Fare.apply(np.ceil)
```

```python
train.head()
```

```python
train.Name=train.Name.str.split(",")
```

```python
train.head()
```

```python
def get_element(mylist,position):
    return mylist[position]
    
```

```python
train.Name.str.split(",").apply(get_element,position=0).head()
```

```python
train.Name.str.split(",").apply(lambda x: x[0]).head()
```
